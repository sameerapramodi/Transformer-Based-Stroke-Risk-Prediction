```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Scale Age
scaler = StandardScaler()
data['Age_scaled'] = scaler.fit_transform(data[['Age']])

# Create Age groups
bins = [0, 18, 30, 40, 50, 60, 70, 80, 100]
labels = ['<18','18-29','30-39','40-49','50-59','60-69','70-79','80+']
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Encode Gender
le = LabelEncoder()
data['Gender_encoded'] = le.fit_transform(data['Gender'])
# Male = 1, Female = 0

# Encode Residence Type
data['Residence_encoded'] = le.fit_transform(data['Residence Type'])
# Rural = 0, Urban = 1

display(data[['Age', 'Age_scaled', 'Age_group', 'Gender', 'Gender_encoded', 'Residence Type', 'Residence_encoded']].head())
```

```python
columns_to_drop = ['Age', 'Gender', 'Marital Status', 'Residence Type', 'Work Type']
data = data.drop(columns=columns_to_drop, errors='ignore')

display(data.head())
```

```python
# One-hot encode 'Smoking Status' and 'Alcohol Intake'
data = pd.get_dummies(data, columns=['Smoking Status', 'Alcohol Intake'], drop_first=True, dtype=int)

# Display the first few rows to show the new columns
display(data.head())
```

```python
# Function to extract Systolic and Diastolic pressure
def extract_blood_pressure(bp_string):
    if isinstance(bp_string, str) and '/' in bp_string:
        systolic, diastolic = bp_string.split('/')
        return int(systolic), int(diastolic)
    return None, None

# Apply the function to create new columns
data['Systolic Pressure'], data['Diastolic Pressure'] = zip(*data['Blood Pressure Levels'].apply(extract_blood_pressure))

# Function to extract HDL and LDL cholesterol
def extract_cholesterol(cholesterol_string):
    if isinstance(cholesterol_string, str) and 'HDL:' in cholesterol_string and 'LDL:' in cholesterol_string:
        hdl_part = cholesterol_string.split('HDL: ')[1].split(',')[0]
        ldl_part = cholesterol_string.split('LDL: ')[1]
        return int(hdl_part.strip()), int(ldl_part.strip())
    return None, None

# Apply the function to create new columns
data['HDL Cholesterol'], data['LDL Cholesterol'] = zip(*data['Cholesterol Levels'].apply(extract_cholesterol))

# Display the new columns
display(data[['Blood Pressure Levels', 'Systolic Pressure', 'Diastolic Pressure', 'Cholesterol Levels', 'HDL Cholesterol', 'LDL Cholesterol']].head())
```

```python
# Drop the original 'Blood Pressure Levels' and 'Cholesterol Levels' columns
data = data.drop(columns=['Blood Pressure Levels', 'Cholesterol Levels'])

# Display the first few rows of the updated data DataFrame
display(data.head())
```

```python
# One-hot encode 'Physical Activity'
data = pd.get_dummies(data, columns=['Physical Activity'], drop_first=True, dtype=int)

# Display the first few rows to show the new columns
display(data.head())
```
